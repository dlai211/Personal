```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import plotly.express as px
import pandas as pd
```

## Changes: spent 1,556,000 guild fund in level 2 Emperium

## Current Data I Collected

```python
guild_fund = [6.091e7, 6.248e7, 6.388e7, 6.595e7, 6.771e7, 6.932e7, 7.127e7, 7.357e7, 7.586e7]
guild_fund_cont = [6.202e7, 6.372e7, 6.530e7, 6.694e7, 6.852e7, 7.062e7, 7.267e7, 7.4e7, 7.545e7, 7.684e7]

guild_fund_fit = [i-1.556e7 for i in guild_fund]
guild_fund_fit = guild_fund_fit + guild_fund_cont
guild_fund = guild_fund + guild_fund_cont

diff = np.diff(guild_fund).tolist()
diff_fit = np.diff(guild_fund_fit).tolist()
diff.insert(0,1.2e6)
diff_fit.insert(0,1.2e6)
num_day = len(guild_fund)

data = {'Date': pd.date_range("3/19/2022", periods=num_day), 'Guild Fund': guild_fund, 'Daily Gain': diff}
data_fit = {'Date': pd.date_range("3/19/2022", periods=num_day), 'Guild Fund': guild_fund_fit, 'Daily Gain': diff_fit}
df = pd.DataFrame(data, index=None)
df_fit = pd.DataFrame(data_fit, index=None)
df_fit
```

## Visualizing Sample Guild Fund and Daily Gain 

```python
pd.options.plotting.backend = "plotly"
#fig = px.line(df, x="year", y="lifeExp", color='country')
px.line(df, x='Date', y='Guild Fund', text='Daily Gain', markers=True, title='Genesis Guild Fund Record')
```

## Plotting the Best Fit Line (linear regression)

```python
px.scatter(df_fit, x="Date", y="Guild Fund", text='Daily Gain', trendline="ols", trendline_scope="overall", title='Guild Fund Fitted Line')
```

## A Table of Predicted Guild Fund and Predicted Daily Gain (using data the latest week).

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(df_fit['Date'][-7:-1].values.reshape(-1,1), df_fit['Guild Fund'][-7:-1])

print(f'Model slope: {reg.coef_[0]}, Model intercept: {reg.intercept_}')
date_pred = pd.date_range("04/01/2022", periods=30)
y_pred = reg.predict(date_pred.values.astype(float).reshape(-1,1))
diff_pred = np.diff(y_pred).tolist()
diff_pred = np.ones(len(y_pred))*diff_pred[0]

pred = pd.DataFrame({'Predicted Date': date_pred, 'Predicted Guild Fund': y_pred, 'Predicted Daily Gain': diff_pred})
pred
```

## The Date we will Reached the Goal (Reach lvl 3 Guild)

```python
goal = 9.459e7
pred.iloc[pred[pred['Predicted Guild Fund'].gt(goal)].index[0]]
```
